```python
penguins['bill_length_mm'] = penguins['bill_length_mm'].fillna(penguins['bill_length_mm'].mean())
penguins['bill_depth_mm'] = penguins['bill_depth_mm'].fillna(penguins['bill_depth_mm'].mean())
penguins['flipper_length_mm'] = penguins['flipper_length_mm'].fillna(penguins['flipper_length_mm'].mean())
penguins['body_mass_g'] = penguins['body_mass_g'].fillna(penguins['body_mass_g'].mean())
penguins['sex'] = penguins['sex'].fillna(penguins['sex'].mode()[0])
```

<h2>>Question No.2</h2>
<h3>Is the weight of a penguin more influenced by its sex or by its species?</h3>

```python
ad_f_avg = np.round(adelie_df[adelie_df['sex'] == 'female']['mass_kg'].mean())
ad_m_avg = np.round(adelie_df[adelie_df['sex'] == 'male']['mass_kg'].mean())
print("Adelie   - Male:", ad_m_avg, "kg" "| Female:", ad_f_avg, "kg")
```

```python
ch_f_avg = np.round(chinstrap_df[chinstrap_df['sex'] == 'female']['mass_kg'].mean())
ch_m_avg = np.round(chinstrap_df[chinstrap_df['sex'] == 'male']['mass_kg'].mean())
print("Chinstrap   - Male:", ch_m_avg, "kg" "| Female:", ch_f_avg, "kg")
```

```python
gen_f_avg = np.round(gentoo_df[gentoo_df['sex'] == 'female']['mass_kg'].mean())
gen_m_avg = np.round(gentoo_df[gentoo_df['sex'] == 'male']['mass_kg'].mean())
print("Gentoo   - Male:", gen_m_avg, "kg" "| Female:", gen_m_avg, "kg")
```

<h3>Insights:</h3>
<p>The data shows that a penguin’s weight depends more on its species than its sex. For example, a female Gentoo is still heavier than a male Adelie, which means species matters more than whether it’s male or female.
</p>

<h2>>Question No.5</h2>
<h3>Do penguins from different islands show differences in weight?</h3>

```python
biscoe_mass = np.round(penguins[penguins['island'] == 'Biscoe']['mass_kg'].mean())
dream_mass = np.round(penguins[penguins['island'] == 'Dream']['mass_kg'].mean())
torgersen_mass = np.round(penguins[penguins['island'] == 'Torgersen']['mass_kg'].mean())

print("Biscoe     - Mass:", biscoe_mass, "kg")
print("Dream      - Mass:", dream_mass, "kg")
print("Torgersen  - Mass:", torgersen_mass, "kg")
```

<h3>Insights:</h3>
<p>Penguins from Biscoe Island have the highest average mass, while those from Torgersen are the smallest, suggesting that penguin weight vary depending on their island habitat. Probably because Biscoe Island is more abundant in food resource.</p>
